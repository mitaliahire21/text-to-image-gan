Text Preprocessing Code

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

def preprocess(text):
    tokens = word_tokenize(text.lower())
    tokens = [w for w in tokens if w.isalnum()]
    return tokens

text = "A small yellow flower in a garden!"
print(preprocess(text))
```

Text Embedding

```python
from transformers import BertTokenizer, BertModel
import torch

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt")
    outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1)

embedding = get_embedding("small yellow flower")
print(embedding.shape)
```

Simple Generator

```python
import torch.nn as nn

class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Sequential(
            nn.Linear(100 + 768, 256),
            nn.ReLU(),
            nn.Linear(256, 784),
            nn.Tanh()
        )

    def forward(self, noise, text_embed):
        x = torch.cat((noise, text_embed), dim=1)
        return self.fc(x)

generator = Generator()
```

generate image

```python
noise = torch.randn(1, 100)
fake_image = generator(noise, embedding)

import matplotlib.pyplot as plt

image = fake_image.detach().numpy().reshape(28, 28)
plt.imshow(image)
plt.axis('off')
plt.show()
```
